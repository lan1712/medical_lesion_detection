```python
import zipfile
import os

# Giải nén file dữ liệu
with zipfile.ZipFile('X Ray chest.v1i.yolov8.zip', 'r') as zip_ref:
    zip_ref.extractall('/content/dataset')

print("✅ Đã giải nén xong vào thư mục /content/dataset")
```

```python
from ultralytics import YOLO

# Load mô hình
model = YOLO('yolov8n.pt')

# Train
model.train(
    data='/content/dataset/data.yaml', # Nhớ kiểm tra file này có trong thư mục dataset chưa
    epochs=50,
    imgsz=640,
    batch=16,
    device=0, # BẮT BUỘC  dùng GPU T4
    name='chest_xray_final'
)
```
